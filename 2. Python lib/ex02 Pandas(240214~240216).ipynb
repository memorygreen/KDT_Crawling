#### 1. Series 생성

```python
# Series 생성
h = pd.Series([179, 165, 140, 158, 192])
h
#출력:
# 0    179
# 1    165
# 2    140
# 3    158
# 4    192
# dtype: int64
```

```python
#글자 인덱스 부여해서  Series 생성
w = pd.Series([60,58,81,77,47], index = ["선영표","정봉균","황해도",
                                        "박병관", "정형"])
w
#출력:
# 선영표    60
# 정봉균    58
# 황해도    81
# 박병관    77
# 정형     47
# dtype: int64
```

```python
# 이름을 부여해서 Series 생성
w = pd.Series([60,58,81,77,47], index = ["선영표","정봉균","황해도",
                                        "박병관", "정형"], name="몸무게")
w
#출력:
# 선영표    60
# 정봉균    58
# 황해도    81
# 박병관    77
# 정형     47
# Name: 몸무게, dtype: int64
```

```python
# 속성 활용하기
display(w.shape) # 모양 확인 #출력:(5,)
display(w.values) #값을 꺼내고 싶은 경우  #출력:array([60, 58, 81, 77, 47], dtype=int64)
display(w.index) #인덱스만 꺼내고 싶은 경우 #출력:Index(['선영표', '정봉균', '황해도', '박병관', '정형'], dtype='object')
display(w.dtype) #데이터 타입을 알고 싶은 경우 #출력:dtype('int64')
```

```python
#연산->같은 인덱스끼리 요소별 연산지원
# 같은 인덱스가 없는 경우 연산할 수 없어 NaN 발생
w / (h ** 2)
#출력:
# 0     NaN
# 1     NaN
# 2     NaN
# 3     NaN
# 4     NaN
# 박병관   NaN
# 선영표   NaN
# 정봉균   NaN
# 정형    NaN
# 황해도   NaN
# dtype: float64
```

```python
#키 데이터(h)에 인덱스를 부여
h.index = ["선영표","정봉균","황해도","박병관", "정형"]
#자 : 이미 생성된 시리즈 객체면 속성으로 접근

h

#출력:
# 선영표    179
# 정봉균    165
# 황해도    140
# 박병관    158
# 정형     192
# dtype: int64
```

```python
# cm-> m 단위로 변경 후 계산
w / ((h/100) ** 2)
#출력:
# 선영표    18.726007
# 정봉균    21.303949
# 황해도    41.326531
# 박병관    30.844416
# 정형     12.749566
# dtype: float64
```

```python
h/100
#출력
# 선영표    1.79
# 정봉균    1.65
# 황해도    1.40
# 박병관    1.58
# 정형     1.92
# dtype: float64
```

```python
# 인덱싱 & 슬라이싱
display(h["선영표"]) #출력 : 179
display(h[0]) #출력 : 179
display(h[["선영표","정봉균"]])
#출력:
# 선영표    179
# 정봉균    165
# dtype: int64
display(h[[0,3]])
#출력:
# 선영표    179
# 박병관    158
# dtype: int64



#자: 글자 슬라이싱도 가능
display(w["선영표":"박병관"]) #글자 슬라이싱은 끝 인덱싱을 포함
#출력:
# 선영표    60
# 정봉균    58
# 황해도    81
# 박병관    77
# Name: 몸무게, dtype: int64

#슬라이싱
display(w[1:4])# 숫자 슬라이싱은 끝 인덱스 직전까지 잘라온다
#출력:
# 정봉균    58
# 황해도    81
# 박병관    77
# Name: 몸무게, dtype: int64
```

```python
# boolean 색인
java = pd.Series({"선영표":90,"박병관":79,"정봉균":57,"정형":99})
java
#출력:
# 선영표    90
# 박병관    79
# 정봉균    57
# 정형     99
# dtype: int64

java[java <= 80]
#출력:
# 박병관    79
# 정봉균    57
# dtype: int64
```

```python
#조건을 여러 개 넣고 싶은 경우 -> &(and), |(or) 연산자를 이용
java[(java >= 90)|(java <= 60)]
# 출력:
# 선영표    90
# 정봉균    57
# 정형     99
# dtype: int64    
```

```python
# 데이터 로딩
crime15=pd.read_csv("./data/2015.csv", encoding='euc-kr')
crime16=pd.read_csv("./data/2016.csv", encoding='euc-kr')
crime17=pd.read_csv("./data/2017.csv", encoding='euc-kr')
```

```python
#head()
#자:다행히 컬럼명이 같다~확인가능!
display(crime15.head())
display(crime16.head())
display(crime17.head())
```

```python
# 데이터 필터링(불리언 색인) : 구분 컬럼 값이 '발생건수'와 같은 데이터 추출
crime15_ft = crime15[crime15["구분"]=="발생건수"].copy()
crime15_ft
```

```python
#자:필터링한 데이터에서 집계내리기(합계 sum)
# 가로합게=>axis=1
crime15_ft.loc[:,'살인':'폭력'].sum(axis = 1)

#출력
# 0     18830
# 6      2355
# 12     4720
# 18     2117
# 24     5466
# 30     4172
# dtype: int64
```

```python
crime15_ft['2015총계']=crime15_ft.loc[:,'살인':'폭력'].sum(axis = 1)
crime15_ft
```

```python
#2016
crime16_ft = crime16[crime16["구분"]=="발생건수"].copy()
crime16_ft['2016총계']=crime16_ft.loc[:,'살인':'폭력'].sum(axis = 1)
crime16_ft
```

```python
#2017
crime17_ft = crime17[crime17["구분"]=="발생건수"].copy()
crime17_ft['2017총계']=crime17_ft.loc[:,'살인':'폭력'].sum(axis = 1)
crime17_ft
```

```python
# 17년에만 등장하는 행(관서명 : 광주지방경찰청 =>인덱스 6번) 삭제
crime17_ft.drop(6, inplace=True)
```

```python
crime15_ft
```

```python
# 올바른 계산을 위해 인덱스를 맞춰추자
#1. 데이터프레임.index = 새로운인덱스 or 데이터프레임.columns = 새로운 컬럼즈
#2. 데이터프레임.rename() 함수 활용
#3. 기존 컬럼을 인덱스로 설정 ->set_index()
crime15_ft.set_index("관서명", inplace=True)
crime16_ft.set_index("관서명", inplace=True)
crime17_ft.set_index("관서명", inplace=True)
```

```python
# 증감율 계산
#(금년-작년)/작년 *100

crime15to16 = (crime16_ft['2016총계']- crime15_ft['2015총계'])\
                /crime15_ft['2015총계']*100

crime16to17 = (crime17_ft['2017총계']- crime16_ft['2016총계'])\
                /crime16_ft['2016총계']*100
```

```python
#병합 
result = pd.concat([crime15_ft['2015총계'],crime15to16,crime16_ft['2016총계'],\
                   crime16to17,crime17_ft['2017총계']], axis=1)
#자:열이름 바꾸기
result.rename(columns={0:"2015-2016증감율", 1:"2016-2017증감율"}, inplace=True)
result
```
