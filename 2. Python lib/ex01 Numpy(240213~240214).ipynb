```python
# 육군신체정보를 이용해 BMI 지수를 구해보자
# 데이터 로딩
data = np.loadtxt("./data/육군 신체측정정보2.csv", delimiter=",")
data.shape #출력:(99425, 10)
```

```python
# 키와 몸무게 열을 확인 후 약 10만명에 대한 BMI를 구해보자

#자영풀이
# 신장(단위 : m) 열 추출
height_cm=data[:,4].reshape(-1,1) # 신장 열 추출, 보기 좋게 2차원으로 모양 변경
height_cm

#출력:
# array([[185.5],
#        [167.2],
#        [179.9],
#        ...,
#        [169.9],
#        [165.7],
#        [183.4]])
```

```python
#신장 단위 변환 (cm -> m)
height_m = height_cm*0.01
height_m
#출력:
# array([[1.855],
#        [1.672],
#        [1.799],
#        ...,
#        [1.699],
#        [1.657],
#        [1.834]])
```

```python
#체중 열 추출
weight=data[:,9].reshape(-1,1) # 체중 열 추출, 보기 좋게 2차원으로 모양 변경
weight
#출력:
# array([[73.2],
#        [65.1],
#        [93.2],
#        ...,
#        [87.4],
#        [85. ],
#        [83.4]])
```

```python
# 요소별 BMI 계산
BMI=weight/height_m*height_m
BMI
#출력:
# array([[73.2],
#        [65.1],
#        [93.2],
#        ...,
#        [87.4],
#        [85. ],
#        [83.4]])
```

```python
#자영풀이2-1 (열 합쳐서 하나의 데이터 SET으로 만들기)
#np.concatenate 시도 ->실패 =>해결 : 합치고 싶은 데이터를 튜플로 묶어야함! 축 axis 차원 주의!)
data_new1=np.concatenate((data[:,4],data[:,9]), axis=0)
data_new1#출력:array([185.5, 167.2, 179.9, ...,  87.4,  85. ,  83.4])
```

```python
#자영풀이2-2 (열 합쳐서 하나의 데이터 SET으로 만들기)
#np.append
data_new2=np.append(data[:,4].reshape(-1,1),data[:,9].reshape(-1,1), axis=1)
data_new2
#출력
# array([[185.5,  73.2],
#        [167.2,  65.1],
#        [179.9,  93.2],
#        ...,
#        [169.9,  87.4],
#        [165.7,  85. ],
#        [183.4,  83.4]])
```

```python
#자영풀이2-3 (열 합쳐서 하나의 데이터 SET으로 만들기)
#hstack함수 (자영:h는 수평방향! stack 함수와 다르게 axis 필요없음)
data_new3=np.hstack((data[:,4].reshape(-1,1),data[:,9].reshape(-1,1)))
data_new3
#출력
# array([[185.5,  73.2],
#        [167.2,  65.1],
#        [179.9,  93.2],
#        ...,
#        [169.9,  87.4],
#        [165.7,  85. ],
#        [183.4,  83.4]])
```

```python
#신장데이터 (단위 cm->m 환산)
h = data_new3[:,:1]*0.01#신장
h
# 출력
# array([[1.855],
#        [1.672],
#        [1.799],
#        ...,
#        [1.699],
#        [1.657],
#        [1.834]])
```

```python
w = data_new3[:,1:] #체중
w
#출력
# array([[73.2],
#        [65.1],
#        [93.2],
#        ...,
#        [87.4],
#        [85. ],
#        [83.4]])
```

```python
#bmi
data_bmi = w / h **2
data_bmi
#출력
# array([[21.27273124],
#        [23.28672649],
#        [28.79742034],
#        ...,
#        [30.27782503],
#        [30.95807221],
#        [24.795187  ]])
```

```python
#bmi추가
#concatenate함수 사용
data_con = np.concatenate((data_new3 , data_bmi), axis=1)
data_con

#출력
# array([[185.5       ,  73.2       ,  21.27273124],
#        [167.2       ,  65.1       ,  23.28672649],
#        [179.9       ,  93.2       ,  28.79742034],
#        ...,
#        [169.9       ,  87.4       ,  30.27782503],
#        [165.7       ,  85.        ,  30.95807221],
#        [183.4       ,  83.4       ,  24.795187  ]])
```
