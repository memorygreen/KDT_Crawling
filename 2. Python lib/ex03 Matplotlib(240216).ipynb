#### 0.matplotlib 패키지 import

```python
import matplotlib.pyplot as plt
#한글폰트 적용
plt.rcParams['font.family'] ='Malgun Gothic'
# 마이너스 기호 적용
plt.rcParams['axes.unicode_minus'] =False
```

#### 1. matplotlib를 사용하는 순서
- 1. 그림이 그려질 그림판을 셋팅
- 2. 실제로 그려질 차트 설정
- 3. 그림을 그려달라는 명령어

#### 2. 선 그래프


```python
#자:그림판 크기 설정
plt.figure(figsize=(10,5)) # 가로, 세로(인치)

#자:데이터
x = [1,2,3]
y = [1,2,3]

#자:그림그리는 함수 plot()
plt.plot(x,y, ls=":", # ls - / -- / -. / : #자:ls = line style
            c="red",
            lw=5,
            marker="o", #marker o / ^ / * / + / >
            mec='green',
            ms=20,
            mew=4)
#자:그림 그려달라
plt.show()
```

```python
# 여러개를 차트에 그리기
plt.figure(figsize=(10,5))

plt.plot([1,2,3], [1,2,3], label="인구수")
plt.plot([1,2,3], [1,4,9], label="생산량")

#범례
plt.legend()

plt.show()
```

#### 3. bar 차트


```python
plt.figure(figsize=(10,5))

#bar 차트 ->범주형 데이터와 수치형 데이터를 그릴 대 
x = ['A형', 'B형', 'AB형', 'O형']
y = [100000, 5, 7, 3]

plt.bar(x, y)
plt.ylim(0,10)
plt.xlabel("혈액형")
plt.ylabel("인원수")
plt.title("우리반 혈액형 분포")
plt.grid()

plt.show()
```

#### CCTV 데이터 실습
- 행정동별 총 카메라 수를 집계하고 bar차트로 시각화해보자

```python
import pandas as pd
```

```python
cctv = pd.read_csv("./data/광주광역시_CCTV_20231231.CSV", encoding='euc-kr')
cctv.head()
```

```python
#자영 풀이
# 소재지지번주소 열에서 -> 행정동 만 추출하여 => '행정동'열 추가
cctv['행정동']=cctv['소재지지번주소'].str.split(" ").str[1].str.strip()
cctv
```

```python
#행정동열을 기준으로 모든 열의 합 구하기
dong_sum=cctv.groupby("행정동").sum()
```

```python
dong_sum
```

```python
#막대그래프의 y축에 사용할 
#'카메라대수' 의 합계만 추출하기
dong_sum['카메라대수']

#출력:
#행정동
# 각화동     55
# 건국동      2
# 계림동    169
# 고룡동      3
# 광산동      6
#       ... 
# 화장동      2
# 화정동    356
# 황금동      8
# 효령동     14
# 흑석동     41
# Name: 카메라대수, Length: 182, dtype: int64
```

```python
# 막대그래프의 x축에 사용할 (범주형데이터) 행정동명 구하기
dong_sum['카메라대수'].index

#출력:
# Index(['각화동', '건국동', '계림동', '고룡동', '광산동', '광천동', '구동', '구소동', '궁동', '금곡동',
#        ...
#        '하남동', '하산동', '학동', '행암동', '호남동', '화장동', '화정동', '황금동', '효령동', '흑석동'],
#       dtype='object', name='행정동', length=182)
```

```python
dong_sum['x'] = 0
```

```python
dong_sum['xx'] = " "
```

```python
#막대그래프 그리기
#1. 공간 만들기
plt.figure(figsize=(100,5))

#x,y 축 값 변수로 빼두기
x = dong_sum['카메라대수'].index
y = dong_sum['카메라대수']

#차트 그리기
plt.bar(x, y)

#차트 출력
plt.show()
```

```python
#막대그래프 
#1. 공간 만들기
plt.figure(figsize=(200,5))

x = dong_sum['카메라대수'].index
y = dong_sum['카메라대수']
yy = dong_sum[['카메라대수','x']].values.reshape(-1)
xx = dong_sum.reset_index()[['행정동','xx']].values.reshape(-1)

plt.bar(range(len(x)*2),yy)
plt.xticks(range(len(x)*2), xx)

plt.show()
```
