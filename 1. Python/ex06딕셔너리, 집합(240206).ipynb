### 딕셔너리
- 데이터를 쌍으로 저장하는 자료구조
- key와 value로 묶어서 저장한다.
- key값으로 데이터를 탐색하기 때문에 데이터 접근 속도가 빠르다
- 단, 순서가 존재하지 않는다는 특징이 있다.


#### 1. 딕셔너리 생성

```python
a = {} # 빈 딕셔너리
b = {"name" : "hd"} #자: 아이템 1개 들어가 있다! 
c = { 1:2, 3:4} #숫자를 이용한 딕셔너리 
```

#### 2.딕셔너리 길이 확인

```python
print(len(a)) # 0
print(len(c)) # 2
```

#### 3.데이터 타입 확인

```python
print(type(a)) #<class 'dict'> #모두 객체!
```

#### 4. 딕셔너리에 값 추가

```python
a['name'] = 'hhd' #값 추가시 key와 value를 작성
a #출력: {'name': 'hhd'}
```

#### 5. 값 가져오기

```python
print(a['name']) #인덱스 위치에 key 값을 작성 #출력 : hhd
#print(a['age']) #없는 키 값을 작성하는 경우 예외 발생
print(a.get('age')) #없는 키 값을 작성하는 경우 None 자:get함수 #출력:None
```

#### 6. 딕셔너리 유용한 함수

```python
song = {"노래제목" : "Love wins all",
       "가수" : "아이유",
       "앨범유형" : "싱글/EP",
       "장르": "가요/발라드"}
```

```python
#key 값들만 꺼내고 싶은 경우
song.keys() #출력: dict_keys(['노래제목', '가수', '앨범유형', '장르'])
```

```python
#value 값들만 꺼내고 싶은 경우
song.values() #출력:dict_values(['Love wins all', '아이유', '싱글/EP', '가요/발라드'])
```

```python
 #for문이랑 함께 쓰는 경우 ->key 값을 중심으로 동작
for data in song :
    print(data)
```

```python
# in / not in 을 딕셔너리에 쓰는 경우 -> key값을 중심으로 동작
print("가수" in song) #출력 : True
print("아이유" in song) #출력 : False
```

```python
# 딕셔너리에 각 아이템을 접근하고 싶은 경우
# 주의) 순서가 항상 똑같다는 보장 x
for k,v in song.items():
    print(k, v)
```

```python
song.items() #출력 : dict_items([('노래제목', 'Love wins all'), ('가수', '아이유'), ('앨범유형', '싱글/EP'), ('장르', '가요/발라드')])
```

```python
# 전체 데이터 삭제
song.clear()
```

```python
song #출력: {}
```

### 집합
- 집합연산을 할 수 있는 데이터타입
- 합집합, 차집합, 교집합 연산이 가능
- 순서가 없다
- 중복을 허용하지 않는다.

### 1.집합의 생성(선언)

```python
my_set = set()
my_set #출략:set()
```

```python
my_set2 = set([1,2,3,4,5])
my_set2 #출력:{1, 2, 3, 4, 5}
```

```python
my_set3 = set(['a','p','p','l','e'])
my_set3 #출력:{'a', 'e', 'l', 'p'}
```

```python
my_set4 = set([3,4,7,8,9])
my_set4#출력:{3, 4, 7, 8, 9}
```

### 2. 합집합

```python
my_set2 | my_set4 #출력 : {1, 2, 3, 4, 5, 7, 8, 9}
```

### 3. 교집합

```python
my_set2 & my_set4 #출력 : {3, 4}
```

### 4.차집합

```python
my_set2 - my_set4 # 2번 집합만 고유하게 가진 내용 출력 #출력 : {1, 2, 5}
```

```python
my_set4 - my_set2 #4번 집합만 고유하게 가진 내용 출력 #출력 : {7, 8, 9}
```
