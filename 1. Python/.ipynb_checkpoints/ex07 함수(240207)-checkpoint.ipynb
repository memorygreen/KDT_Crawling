### 함수
- 지정된 코드에 이름을 붙여 관리하는 방법
- 코드의 간결함을 높여주고, 재사용성을 높여준다.
- python은 OOP도 지원하기 때문에 메소드 개념도 동시에 존재한다.


```python
#함수 정의
def number_sum(n1, n2) : 
    result = n1 + n2
    return result
```

```python
#함수 호출
rs = number_sum(2,10)
print(rs) #출력 : 12
```

#### 1번 문제


```python
#자영 풀이
def number_sub1(num1, num2):
    result = num1 - num2
    return result
```

```python
#두 수를 입력 받아서 뺀 결과를 return하는 함수를 정의하시오
num1 = int(input('첫 번째 정수 입력>>'))
num2 = int(input('두 번째 정수 입력>>'))

# 빼기 기능을 하는 함수 두 인자값 입력하여 호출
result = number_sub1(num1, num2)
print(result)
```

```python
#쌤풀이
def number_sub(num1, num2):
    return num1 - num2
```

```python
# 쌤풀이
num1 = int(input('첫 번째 정수 입력>>'))
num2 = int(input('두 번째 정수 입력>>'))
rs = number_sub(num1, num2)
print(rs)
```

#### 2번 문제


```python
#자영풀이
def divisor1(num):
    for i in range(1, num+1, 1):
        if num % i == 0 :
            print(i, end=" ")
        
```

```python
#자영풀이
num = int(input("숫자를 입력하세요>>"))
divisor1(num)
```

```python
#쌤풀이
def divisor(input_number):
    for i in range(1, input_number+1):
        if input_number % i == 0:
            print(i, end=" ")
```

```python
divisor(32)#출력:1 2 4 8 16 32 
```

```python
divisor(100) #출력:1 2 4 5 10 20 25 50 100 
```

### 3. 파이썬 함수 특징

### 1. 전역변수와 지역변수

```python
# 함수에서 선언된 변수는 지역변수다
a = 10 #전역변수 선언
def test() :
    a = 1 # 지역변수로 인식

test()
print(a)
```

```python
# global 키워드를 이용해 전역변수로 인식시키자
a = 10 #전역변수 선언
def test() :
    global a
    a = 1 # 전역변수로 인식

test()
print(a)
```

### 2. 매개변수에 기본값 설정
- 매개변수에 기본값을 설정하여 호출시 인자를 생략할 수 있다

```python
def hello(name, age):
    print(f"{name},{age}님 안녕하세요.")
```

```python
hello('홍길동',50) #출력:홍길동,50님 안녕하세요.
```

```python
def hello(name, age=10): #
    print(f"{name},{age}님 안녕하세요.")
```

```python
hello('홍길동') #출력:홍길동,10님 안녕하세요.
```

### 3. 리턴 값을 여러 개 줄 수 있다
- retrun 할 때 여러 값을 튜플로 묶어서 반환
- 호출시에 여러 변수를 만들어서 받을 수 있음


```python
def getData() : 
    return 'A', 10
```

```python
a,b = getData() #출력:('A', 10) # 자:사실 1개데이터 리턴하지만 편의상 2개처럼 보임
print(a)  #출력:A
print(b)  #출력:10
```

#### 4. docstring
- 함수에 대한 부가 설명을 작성하는 기능

```python
def cal(num1 ,num2, op) : 
    """ 덧셈과 뺄셈을 계산하는 함수
        op : 연산자를 의미, 주로 '+' 기호를 입력
        
    """
    if op == '+' :
        return num1 + num2
    else:
        return num1 - num2
```

```python
cal(10, 5, '+') #출력:15
```

##### 5. 가변인자
- 사용자가 입력인자를 정해진 길이 없이 넣고 싶은 경우

```python
#자:튜플 스타일
def fun1(name, *score) :
    print("이름 : ", name)
    print("성적 : ", score)
```

```python
#자:튜플 스타일
fun1("김가가", 100, 80, 40, 50)
#출력
#이름 :  김가가
#성적 :  (100, 80, 40, 50)
```

```python
#자:딕셔너리 스타일
def fun2(name, **score) : 
    print("이름 : ", name)
    print("성적 : ", score)
```

```python
#자:딕셔너리 스타일
fun2("김나나", java=100, python=80, html=40, css=50)
#출력:
#이름 :  김나나
#성적 :  {'java': 100, 'python': 80, 'html': 40, 'css': 50}
```

#### 4. 익명함수

```python
# 일반적인 함수 정의
def mul(x,y):
    return x * y
result = mul(3,5)
print(result) #출력 : 15
```

```python
# 람다식을 이용한 함수 정의
mul2 = lambda x,y : x*y
mul2(9,7)#출력 : 63
```

#### 5. 모듈과 패키지
- 모듈 : 관련된 변수, 함수, 클래스 등을 하나의 파일로 묶어 놓은 것
- 패키지 : 모듈을 여러 개 묶은 것
- 라이브러리 : 패키지를 정리해서 만든 것
