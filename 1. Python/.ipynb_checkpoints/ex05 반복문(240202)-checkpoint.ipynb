### 반복문(while문, for문)
- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문

#### 1. while문
- 반복 횟수가 명확하지 않을 때
```python
while 조건식:
    코드블록
    코드블록
```

````python
# while문을 활용해서 "파이썬 최고!!"를 13번 출력

cnt = 1
while  cnt <= 13:
    print(cnt, "파이썬 최고!!")
    cnt += 1
````

````python
# break - 반복문을 나가는 기능!
while True : 
    print("무한루프")
    break
````

````python
# 자:while문 & if문 & break 써서 횟수 반복

cnt = 1
while  True:
    print(cnt, "파이썬 최고!!")
    cnt += 1
    if cnt > 13:
        break
    
````

````python
# 두 개의 정수를 입력 받아서 더하는 코드를 작성
# 단, 0과 0을 입력했을 때 프로그램 종료 출력!!

while True :
    
    num1 = int(input("첫 번재 정수입력"))
    num2 = int(input("두 번재 정수입력"))
    
    if num1 != num2 :    
        print(f"두 정수의 합>>{num1+num2}")
    
    else: 
        print("프로그램이 종료되었습니다")
        break
````

````python
#쌤풀이

while True:
   
    n1 = int(input("첫 번째 정수입력 >>"))
    n2 = int(input("두 번째 정수입력 >>"))
        
    if n1 == 0 and n2 == 0 :
        print("프로그램이 종료되었습니다.")
        break;
    
    print(f"두 정수의 합 >>{n1 + n2}")
````

````python
# 실습
# 다이어트 관리 프로그램
# 1. 현재 몸무게와 목표몸무게를 입력 받고 주차 별 감량 몸무게를 입력 받으세요.
# 2. 목표몸무게를 달성하면 축하한다는 문구를 출력하고 입력을 멈추세요!
now = int(input("현재 몸무게 : "))
goal = int(input("목표 몸무게 : "))
cnt = 0
while now > goal :
    cnt += 1 
    minus = int(input(f"{cnt}주차 감량 몸무게 : "))
    now = now - minus
print(f"{now}kg 달성!! 축하합니다!")
````

````python
#240205 월
#쌤풀이
cw = int(input('현재 몸무게 : '))
tw = int(input('목표 몸무게 : '))

cnt = 1
while cw > tw :
    lw = int(input(f"{cnt}주차 감량 몸무게 : "))
    cnt = cnt + 1
    cw  = cw - lw # 감량한 몸무게 만큼 현재 몸무게에서 뺄셈
    
print(f"{cw}kg 달성!! 축하합니다!!")    
````

#### 2. for문
- 반복 횟수가 명확할 때 활용
    1. 인덱스를 지정해서 반복하는 방법
    2. foreach 스타일로 리스트, 튜플 등 아이템의 갯수에 따라 반복하는 방법


````python
## 1. 인덱스 지정하는 방법
## range 함수 사용
## range(시작값, 끝값, 증가값)
for i in range(1,10,1) : 
    print(i, end=" ") #줄바꿈 제거 #"자: 기본은 \n"
    #출력:1 2 3 4 5 6 7 8 9 
````

````python
for i in range(10, 1, -1) :
    print(i, end=",") #줄바꿈 제거 #출력:10,9,8,7,6,5,4,3,2,
````

````python
## 2. foreach 스타일 반복문
## 주로 아이템이 들어있는 리스트, 튜플, 딕셔너리 등을 반복하고 싶을 때 사용
word_list = ["apple", "banana", "orange"]
for w in word_list : # 순차적으로 리스트의 아이템을 꺼냄
    print(w, end = " ") #출력:apple banana orange 
    
````

````python
## 문자열 타입은 각 글자마다 인덱스가 부여되기 때문에 반복문에서 활용가능
for c in "apple" : 
    print(c, end = " ") #출력 : a p p l e 
````

````python
## 아이템이 2개씩 묶인 튜플 반복하기
student_score = (('hhd',90), ('pbk',88),('jh',50))
for student in student_score :
    print(student)
````

````python
# 반복되는 아이템을 쪼개서 각각 변수에 넣을 수 있다
for name, score in student_score :
    print("name : ",name)
    print("score : ", score)
````

````python
# foreach 스타일에서 인덱스를 같이 쓰고 싶은 경우
item_list = ['A', "B", 'C', "D", "E"]
for index, item in enumerate(item_list) : # enumerate -> item과 index 같이 꺼내주는 함수
    print(index, "번째 : ", item) 
````

### 실습


````python
#실습 #60점이상 합격, 미만 불합격
#자영풀이
score_list = [90, 45, 70, 60, 55]

for index, score in enumerate(score_list):
    if(score >= 60):
        print(index+1, "번 학생은 합격입니다")
    else:
        print(index+1, "번 학생은 불합격입니다")
````

````python
# 쌤풀이1
score_list = [90, 45, 70, 60, 55]
for n, s in enumerate(score_list) :
    if s >= 60 :
        print((n+1), "번째 학생은 합격입니다")
    else :
        print((n+1), "번째 학생은 불합격입니다")
````

````python
# 쌤풀이2
score_list = [90, 45, 70, 60, 55]
for n, s in enumerate(score_list) :
    rs = "합격" if s >= 60 else "불합격"
    print(f"{n+1}번 학생은 {rs}입니다.")
````

### 문제


````python
# 문제 : 1부터 100사이의 숫자 중 3의 배수인 값들의 합을 출력하세요. (정답 : 1683)
# 자영풀이1 range함수로 1~100 뽑고 % 연산자로 3의 배수 뽑기
sum1 = 0
for i in range(1,101,1):
    if i % 3 ==0:
        sum1 = sum1 + i
print(sum1)
````

````python
#자영풀이2 (range 함수로 3의 배수 뽑기)
sum2=0
for i2 in range(3,101,3):
    sum2 = sum2 + i2
print(sum2)
````

````python
# 쌤풀이
# 1번 풀이
sum_3 = 0 # 총합을 구하기 위한 변수
for num in range(1,100,1):
    if num % 3  == 0: #3의 배수를 판별하는 조건문
        sum_3 = sum_3 + num
print("3의 배수 총합 : ", sum_3)  #출력 : 3의 배수 총합 :  1683 
````

````python
# 쌤풀이2
# 2번 풀이
sum_3 = 0 # 총합을 구하기 위한 변수
for num in range(3, 100, 3):
    sum_3 = sum_3 + num
print("3의 배수 총합 : ", sum_3) #출력 : 3의 배수 총합 :  1683 
````
