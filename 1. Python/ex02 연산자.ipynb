### 삼항연산자
- `(강세표, Grave) : 물결(~) 기호 아래에 있음(코드를 마크다운 셀에 기록으로 남기고자 하는 경우 활용!)

``` python
a if 조건식(자:관계연산자, 비교연산자) else b
```

- 조건식이 True인 경우 a(False인 경우 b)
- 이진분류 : 두 개 중 하나로 분류!

````python
score = 60
"합격" if score >= 60 else "불합격" #출력 : '합격'
````
