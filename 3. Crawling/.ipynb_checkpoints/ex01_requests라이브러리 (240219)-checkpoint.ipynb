### Crawling
- 데이터 수집
- 페이지의 정보를 요청 후 응답 받아 필요한 데이터를 수집하는 행위

##### requests 라이브러리
- 서버에게 페이지 정보(데이터)를 요청하고 응답 받기 위한 라이브러리
- 브라우저 대신 

```python
# 라이브러리 불러오기
import requests as req
```

```python
# 네이버 페이지의 정보를 요청하여 응답받기 (requests 라이브러리 활용)
url = 'https://www.naver.com/' #자 : 문자열 타입으로 넣었다!
res = req.get(url) # 페이지정보 가져와줘~
# Response [200] : 성공적으로 페이지 정보를 응답받음
# Response [400번대]: 클라이언트 pc에서 요청상의 문제가 있을 때
# Response [500번대]: 서버pc에서 응답상의 문제가 있을 때
```

```python
# 확인 (네이버 페이지의 정보를 잘 받아왔는지)
res.text

# 요청한 페이지의 html 정보를 확인할 수 있음
# 이 때 우리가 응답받은 데이터의 형태는? > 문자열!!
# 데이터를 효율적으로 추출하기 위해서는 html 코드로 변환(자:하는 작업 필요!)
```

- 페이지 정보가 업데이트 되면 반드시 다시 페이지 정보를 요청하여 응답받아야한다


```python
# '푸바오'를 검색한 (네이버)페이지의 정보를 요청 응답
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%ED%91%B8%EB%B0%94%EC%98%A4'
res = req.get(url)
res.text
```

```python
# 쌤풀이
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%ED%91%B8%EB%B0%94%EC%98%A4'
res = req.get(url)
res.text
#응답받은 데이터 형태 => 문자열 => HTML 코드로 변환 (태그를 통한 데이터 수집)
```

##### BeautifulSoup 라이브러리 ★☆
- 컴퓨터가 알아먹을 수 있는 언어로 변경해주는 라이브러리(파싱/parsing)
- 뷰티풀숩 객체화

```python
# 라이브러리 불러오기
from bs4 import BeautifulSoup as bs 
```

```python
# bs 객체화
# bs(페이지정보, 파싱방법)
soup = bs(res.text, 'lxml')
# f12 개발자모드에서의 페이지정보를 soup이 담고 있음
```

- 네이버 탭 데이터 수집하기


```python
# '이미지' 글자를 추출하기 위해서 태그선택
soup.select('a') 
# 모든 a 태그를 수집
```

```python
soup.select('root') 
```

```python
# 정확한 태그 정보 선택하는 것이 중요!! (구분자, 선택자 등을 활용)
# soup.select('태그이름 선택자기호 선택자이름')
data = soup.select('a.tab')
data
# 여러 개의 데이터이기 때문에 리스트에 담겨 출력
```

```python
# 우리가 필요한 순수한 콘텐츠만을 추출 '이미지'
# 요소.text
data[0].text
#출력 : '뉴스'
```

```python
# 반복문을 통해서 모든 데이터 추출

for i in range(len(data)):
    print(data[i].text)

#출력:
# 이미지
# 뉴스
# 블로그
# 카페
# 지식iN
# 인플루언서
# 동영상
# 쇼핑
# 어학사전
# 지도
# 도서
# 지식백과
# 학술정보
# 전체
# 이미지
# 뉴스
# 블로그
# 카페
# 지식iN
# 인플루언서
# 동영상
# 쇼핑
# 어학사전
# 지도
# 도서
# 지식백과
# 학술정보
```

- 네이버에서 '날씨' 검색 후 온도 수집

```python
#자영풀이
# 페이지정보 요청 후 응답받기
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%82%A0%EC%94%A8'
res = req.get(url)
res.text
```

```python
# bs 객체화
soup = bs(res.text, 'lxml')
soup
```

```python
# 태그 선택
temp = soup.select('div.temperature_text>strong')
temp[0].text
```

```python
# 순수한 콘텐츠 추출
for i in temp:
    print(i.text)
```

```python
#쌤풀이
# 페이지정보 요청 후 응답받기
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%82%A0%EC%94%A8'
res = req.get(url)
res
```

```python
#bs 객체화
soup = bs(res.text, 'lxml')
```

```python
#태그 선택
temp = soup.select('div.temperature_text>strong')
temp
```

```python
#순수한 콘텐츠 추출
temp[0].text
```

 - select, select_one 차이점 ☆★
     - select: 해당되는 모든 요소를 추출 -> 리스트 형태로 출력
     - select_one: 해당되는 요소 중 "가장 상단"에 있는 요소 "한 개만" 추출

```python
ex = soup.select_one('div.temperature_text>strong')
#자 : 리스트 형태 아닌 하나의 요소 (인덱스 할 필요없이 바로 .text사용가능)
ex

#출력:
# <strong><span class="blind">현재 온도</span>11.6<span class="celsius">°</span></strong>
```

```python
ex.text
#출력:
# '현재 온도11.6°'

# select_one 기능을 사용하면 요소 형태로 추출 되기 때문에
# 리스트 인덱싱 해줄 필요 없이 바로 text 속성 사용 가능!
# 깔끔한 코드 작성을 위해 사용(최상단 데이터 하나만 가져와도 될 때)
```
