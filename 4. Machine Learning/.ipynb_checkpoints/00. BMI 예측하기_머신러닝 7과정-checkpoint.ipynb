### 머신러닝 종류
1. 지도학습
    - 문제(X)와 답(y)가 주어진 상태에서 학습
    - 분류, 회귀
        - 분류 : 정답 데이터(y)의 형태가 범주형 데이터일 때 
        - 회귀 : 정답 데이터(y)의 형태가 연속형 데이터일 때 
        
2. 비지도학습
    - 문제(X)에 대해서만 주어진 상태에서 학습
    - 데이터의 숨겨진 특성, 구조, 패턴을 파악 -> 클러스터링(군집화)
    
3. 강화학습
    - 완전한 답을 제공하지 않음
    - 더 큰 보상을 얻을 수 있는 방향으로 학습을 진행
    - ex. 알파고

```python
# 정답데이터 확인하기(label)
data.index.unique()
# 총 6ea의 클래스를 가짐
# 다중분류 : 6가지의 클래스 중에서 1개의 정답을 예측

# Extreame Obesity : 고도 비만
# Obesitry : 비만
# Overweight : 과체중
# Normal : 정상
# Weak : 저체중
# Extremely Weak : 심한 저체중
```

```python
# 시각화 (bmi 레이블별 분표 현황 시각화)
# 산점도 (scatter)
# 고도비만 (Extreme Obesity) 에 대한 산점도 그려보기
EO = data.loc['Extreme Obesity']


EO['Height'] #EO.loc[:,'Height']
EO['Weight'] #EO.loc[:,'Height']

plt.scatter(EO['Height'], EO['Weight'], color = "purple", label='ext ob')

plt.legend()
plt. show()
```

```python
# 같은 작업을 6번 동일하게 진행해서, 모든 클래스에 대한 산점도 그려보기
# 함수 활용!

# 자영풀이
def draw_scaater(cate, col):
    category = data.loc[cate]
    category['Height'] #EO.loc[:,'Height']
    category['Weight'] #EO.loc[:,'Height']
    plt.scatter(category['Height'], category['Weight'], color = col, label=cate)

cate1='Extreme Obesity'
col1 = 'red'
draw_scaater(cate1, col1)

cate2= 'Obesity'
col2 = 'orange'
draw_scaater(cate2, col2)

cate3= 'Overweight'
col3 = 'yellow'
draw_scaater(cate3, col3)

cate4= 'Normal'
col4 = 'green'
draw_scaater(cate4, col4)

cate5= 'Weak'
col5 = 'blue'
draw_scaater(cate5, col5)

cate6= 'Extremely Weak'
col6 = 'purple'
draw_scaater(cate6, col6)


plt.legend()
plt. show()
```

```python
# 쌤풀이
# 같은 작업을 6번 동일하게 진행해서, 모든 클래스에 대한 산점도 그려보기
# 함수 활용!
def graph_draw(label, c) :
    class_bmi = data.loc[label]
    plt.scatter(class_bmi['Height'], class_bmi['Weight'], 
                color = c, label=label)
```

```python
data.index.unique()
```

```python
# 함수 호출하기

graph_draw("Extreme Obesity", 'red')
graph_draw("Obesity",'orange')
graph_draw("Overweight", 'yellow')
graph_draw("Normal",'green')
graph_draw("Weak", 'blue')
graph_draw("Extremely Weak", 'purple')

plt.legend()
plt. show()
```

```python
# 데이터 분리
# 문제 데이터 (X) : Height, Weight
# 정답 데이터(y) : label
X = data[["Height", "Weight"]] # X(문제) 대문자! -> 2차원
y = data.index
```

```python
X.shape, y.shape # ((500, 2), (500,))
```

```python
# train, test 분리
# 한정적인 데이터 내에서 학습 + 평가까지 해야되기 때문에
# train : 70%, test : 30% (ex. 7.5: 2.5)

X_train = X.iloc[:350] #0~349  #500*70%=350
y_train = y[:350]

X_test = X.iloc[350:]
y_test = y[350:]
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape #((350, 2), (150, 2), (350,), (150,))
```
