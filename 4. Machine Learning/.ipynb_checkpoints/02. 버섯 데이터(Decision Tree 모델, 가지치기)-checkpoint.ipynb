### Decision Tree(의사결정 나무)
- 스무고개 하듯이 True/False 질문을 반복하며 학습
- 특정 기준(질문, node)에 따라 데이터를 구분하는 모델
- 분류나 회귀 모두 사용 가능


```python
# 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# 문리하는 도구
from sklearn.model_selection import train_test_split
```

#### 1. 문제정의
- 버섯데이터를 활용해서 독성/식용 버섯을 분류


#### 2. 데이터 수집

```python
data = pd.read_csv("./data/mushroom.csv")
data
```

```python
# 데이터 크기 확인
data.shape # (8124, 23)
# 특정 22ea +  정답 1ea
```

```python
data.info()
# 결측치 없음 -> 별도로 전처리 X
# sklearn 모델들은 문자열 값을 입력으로 허용하지 않는다!
# 데이터 타입 확인 시 object -> int, float 변경
```

```python
data.head(10)
```

```python
# X, y 데이터 분리해서 변수에 담기
X = data.drop('poisonous', axis=1)
y = data['poisonous']
```

```python
X.shape, y.shape
#((8124, 22), (8124,))
```

```python
y.value_counts()
# poisonous
# e    4208
# p    3916
# Name: count, dtype: int64
```

```python
# X의 기술통계량(요약본) 확인
X.describe().T

# count : 데이터 개수
# unique : 교유값 개수
# top : 가장 많은 개수의 문자열
# freq : 가장 많은 개수의 문자열의 빈도수
```

#### 3. 데이터 전처리
- 결측치 대체, 제거
- 이상치 대체, 제거
- 변수처리
    - 인코딩 : 범주형 데이터를 숫자형으로 변환
    - label encoding : 수치값으로 mapping 하는 작업
    - one_hot endocing : 0 or 1 의 값을 가진 여러 개의 새로운 특성으로 변경하는 작업
    
- train, test 분리


```python
# label encoding
# 바뀌어진 숫자값들의 높고 낮음의 의미가 필요할 때 
# 자: ex. 성적
X['cap-shape'].map({'x':0, 'f':1, 'k':2, 'b':3, 's':4, 'c':5 })
# 0       0
# 1       0
# 2       3
# 3       0
# 4       0
#        ..
# 8119    2
# 8120    0
# 8121    1
# 8122    2
# 8123    0
# Name: cap-shape, Length: 8124, dtype: int64
```

```python
# one_endocing
# 바뀌어진 숫자값들의 높고 낮음의 의미가 필요하지 않을 때 
# 데이터의 unique 개수만큼 컬럼을 추가
# 1 or 0 

X_one_hot = pd.get_dummies(X)
X_one_hot
# 데이터 값의 각 unique의 개수만큼 컬럼의 수가 늘어남
# 총 컬럼의 수 22ea -> 117ea
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y,
                                                   test_size=0.3,
                                                    random_state=2024,
                                                    stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
#((5686, 117), (2438, 117), (5686,), (2438,))
```

#### 5. 모델링

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree1 = DecisionTreeClassifier()
tree1.fit(X_train, y_train)

# score() : 모델 자체적으로 제공하는 성능 평가 지표
tree1.score(X_test, y_test) #1.0
```

```python
tree1.score(X_train, y_train) # 1.0
```

### 모델링 + 하이퍼파라미터
- max_depth : 최대 깊이 지정
- max_leaf_nodes : 최대 리프 노드의 수 지정
- max_samples_split : 노드를 분할하기 위한 최소 샘플 수를 지정
    - 지정한 값보다 적은 샘플을 가지는 노드는 분할되지 않음
- max_samples_leaf : 리프노드가 가져야 하는 최소 샘플 수를 지정
    - 분할 전에 지정한 값보다 적은 샘플을 가진 노드가 나오는 경우 분할 방지

```python
tree2 = DecisionTreeClassifier(max_depth = 3)
tree2.fit(X_train, y_train)
tree2.score(X_test, y_test)
#0.9868744872846595
```
